# Handling Irrelevant Values

## orders_dataset

```python
df_orders.columns.to_list()
```

Rename value in order_status to categorize them into delivered, shipped, invoiced, create and canceled, where:
- delivered: the order is delivered to customer
- shipped: the order is on its way to customer, including invoiced, unavailable, processing, and approved
- invoiced: the order is approved by system
- created: customer is in the process of create the order
- canceled: the order is canceled by customer

```python
replace_status_value = ['unavailable','processing','approved']
df_orders[df_orders['order_status'].isin(replace_status_value)]
df_orders_copy = df_orders.copy()
df_orders_copy.replace(to_replace = replace_status_value, value = 'invoiced', inplace=True)
df_orders_copy['order_status'].value_counts()
```

```python
# Because I will analyze success order only, the order that the system have already approved it
# I will drop the data that have null value in order_approved_at column
success_order = df_orders_copy.copy()
success_order.dropna(subset=['order_approved_at'], inplace = True)
success_order.isnull().sum()
```

# Handling Missing Values

## orders_dataset

```python
success_order.isnull().sum()
```

Because the missing data could be the data that have status order incomplete (invoiced, shipped, or canceled) so the data will be displayed and used as it is

```python
#checking data from order_status to see the value_counts of data
success_order['order_status'].value_counts()
```

```python
#In 'canceled' data, the missing value is varied in different columns
success_order[success_order['order_status']=='canceled'].head()
```

#### Separate data frame for each order_status

##### Delivered in order_status

```python
#There should not be any missing value
df_del = success_order[success_order['order_status']=='delivered']
df_del.isnull().sum()
```

Missing value in order_delivered_date column will be deleted.
Missing value in order_delivered_carrier_date and order_approved_at column will be filled with data in their next right column. 

```python
df_del1 = df_del[df_del['order_delivered_customer_date'].notna()]
df_del2 = df_del1.copy()
df_del2['order_delivered_carrier_date'] = df_del2['order_delivered_carrier_date'].fillna(df_del2['order_delivered_customer_date'])
df_del2['order_approved_at'] = df_del2['order_delivered_carrier_date'].fillna(df_del2['order_delivered_carrier_date'])
df_del2.isnull().sum()
```

##### Invoiced in order_status

```python
# There is no null data in order_approved_at so the data will be kept as it is
df_inv = success_order[success_order['order_status']=='invoiced']
df_inv.isnull().sum()
```

##### Shipped in order_status

```python
# There is no null data in order_delivered_carrier_date so the data will be kept as it is
df_ship = success_order[success_order['order_status']=='shipped']
df_ship.isnull().sum()
```

##### canceled in order_status

```python
df_cancel = success_order[success_order['order_status']=='canceled']
df_cancel.isnull().sum()
# The most canceled order is when the order is already delivered to the customer (order_delivered_customer_date) 
```

#### union all data (oreders_dataset)

```python
df_orders1 = pd.concat([df_del2, df_inv,df_ship,df_cancel])
df_orders1.isnull().sum()
```

# Handling outliers

The outliers from every type of payment_type will be checked, so there will be a separate data frame for each of payment_type 

## credit card

```python
df_cc1 = df_merge_all[df_merge_all['payment_type']=='credit_card']
df_cc1
```

```python
df_cc1.boxplot(column=['payment_value'],fontsize=10,
           rot=0,grid=False,figsize=(5,5),vert=False)
```

```python
df_cc1.mean()
```

```python
Q1_cc1 = df_cc1['payment_value'].quantile(0.25)
Q3_cc1 = df_cc1['payment_value'].quantile(0.75)

IQR_cc1 = Q3_cc1 - Q1_cc1
boxplot_min_cc1 = Q1_cc1 - 1.5 * IQR_cc1
boxplot_max_cc1 = Q3_cc1 + 1.5 * IQR_cc1
```

```python
filter_price_min_cc1 = df_cc1['payment_value']<boxplot_min_cc1
filter_price_max_cc1 = df_cc1['payment_value']>boxplot_max_cc1

df_non_outlier_cc1 = df_cc1[~
    (filter_price_min_cc1|filter_price_max_cc1)
    ]

df_non_outlier_cc1['payment_value'].mean()
```

```python
df_non_outlier_cc1.boxplot(column=['payment_value'],fontsize=10,
           rot=0,grid=False,figsize=(5,5),vert=False)
```
